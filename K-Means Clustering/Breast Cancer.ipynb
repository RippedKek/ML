```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
bc = load_breast_cancer()
dir(bc)
```

```python
bc.feature_names
```

```python
bc.target_names
```

```python
df = pd.DataFrame(bc.data, columns = bc.feature_names)
df.head()
```

```python
df['type'] = bc.target
df.head()
```

```python
data = df[['mean radius', 'mean texture']]
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['mean radius'], data['mean texture'])
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Mean Radius vs Mean Texture')
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
yp = km.fit_predict(data)
yp
```

```python
data['prediction'] = yp
data.head()
```

```python
data1 = data[data.prediction==0]
data2 = data[data.prediction==1]
```

```python
plt.scatter(data1['mean radius'], data1['mean texture'], color='blue')
plt.scatter(data2['mean radius'], data2['mean texture'], color='red')
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data[['mean radius', 'mean texture']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng, sse)
plt.xlabel('K')
plt.ylabel('SSE')
```

# Using more than 2 features

```python
df.head()
```

```python
km = KMeans(n_clusters=2)
yp = km.fit_predict(df)
yp
```

```python
df['cluster'] = yp
df.head()
```

```python
df.cluster.unique()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df.drop('cluster', axis='columns'))
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng, sse, color='blue')
plt.scatter(k_rng[1], sse[1], color='red', label='Elbow Point')
plt.legend()
plt.xlabel('K')
plt.ylabel('SSE')
plt.title('K vs SSE (Determining Elbow Point)')
```

### The elbow point is far more defined when used all the features
