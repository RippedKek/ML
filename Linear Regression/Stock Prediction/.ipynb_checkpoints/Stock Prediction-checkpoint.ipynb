```python
# Using daily high price to predict closing price
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
file_path = "C:\\Users\\HP\\stock_data.xlsx"
data = pd.read_excel(file_path)
df = data[['High', 'Close*']] # Using only required columns 
```

```python
plt.scatter(df['High'], df['Close*'], marker='+')
plt.xlabel('Daily High Price')
plt.ylabel('Daily Closing Price')
plt.title('Daily high vs closing')
```

```python
X = df[['High']] # Independent variable
y = df['Close*'] # Dependent on X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

model = LinearRegression()
model.fit(X_train, y_train) # Train the model 
```

```python
y_pred = model.predict(X_test) # Predicted dataset using test dataset 
```

```python
# Plots to finish off the prediction
# Using 50 samples from datasets 
sampleX = X_test.head(50)

# Convert series into dataframe
dfy_test = pd.DataFrame(y_test)
dfy_pred = pd.DataFrame(y_pred)

sampleY = dfy_test.head(50)
samplePred = dfy_pred.head(50)

plt.scatter(sampleX, sampleY, color='blue', marker='+')
plt.plot(sampleX, samplePred, color='red')
```
