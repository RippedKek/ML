```python
# Using daily high and low prices to predict closing price
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
file_path = "C:\\Users\\HP\\Documents\\ML\\Linear Regression\\Stock Prediction\\stock_data.xlsx"
data = pd.read_excel(file_path)
df = data[['High', 'Low', 'Close*', 'Date']] # Create dataframe with only the required columns
```

```python
plt.figure(figsize=(10, 6)) 
plt.bar(df['Date'], df['High'], color='green', label='High')
plt.bar(df['Date'], df['Low'], color='red', label='Low')
plt.ylim(bottom=df['Low'].min() - df['Low'].min()*0.1)

plt.xlabel('Date')
plt.title('High and Low Values Over Time')

# Show the plot
plt.legend()
plt.tight_layout()  
plt.show()
```

```python
X = df[['High', 'Low']] # Independent variables
y = df['Close*']        # Dependent variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 40)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter the test data points
ax.scatter(X_test['High'], X_test['Low'], y_test, color='green', label='True Close*')

# Scatter the predicted data points
ax.scatter(X_test['High'], X_test['Low'], y_pred, color='red', label='Predicted Close*')

# Set labels
ax.set_xlabel('High')
ax.set_ylabel('Low')
ax.set_zlabel('Close*')

# Add a title
plt.title('3D Scatter Plot: High, Low vs Close*')

# Show legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plot the relationship between 'High' and 'Close*'
plt.figure(figsize=(14, 6))

# Scatter plot for High vs Close*
plt.subplot(1, 2, 1)
plt.plot(X_test['High'], y_pred, color='red')
plt.scatter(X_test['High'], y_test, marker='+', color='green', label='True Close*')
plt.xlabel('High')
plt.ylabel('Close*')
plt.title('High vs Close* with Regression Line')
plt.legend()

# Scatter plot for Low vs Close*
plt.subplot(1, 2, 2)
plt.plot(X_test['Low'], y_pred, color='blue')
plt.scatter(X_test['Low'], y_test, marker='+', color='red', label='True Close*')
plt.xlabel('Low')
plt.ylabel('Close*')
plt.title('Low vs Close* with Regression Line')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
high = int(input("Give high value: "))
low = int(input("Give low value: "))
raw_data = {
    "High": [high],
    "Low": [low]
}
new_data = pd.DataFrame(raw_data)
prediction = model.predict(new_data)
print(f"Predicted closing value: {float(prediction[0])}")
categories = ['High', 'Low', 'Predicted Closing Value']

values = [high, low, float(prediction[0])]

plt.bar(categories, values, color=['green', 'red', 'blue'])

plt.ylabel('Value')
plt.title('High, Low, and Predicted Closing Value')

# Show the plot
plt.show()
```
